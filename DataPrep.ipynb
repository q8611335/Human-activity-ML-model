# Data Cleaning - preprocessing

```python
# keep 51 columns needed
df_selected = df_raw[all_needed_cols].copy()
print("Selected data shape:", df_selected.shape)
```

## Filter Female Data

```python
# filter female data
df_female = df_selected[df_selected["_SEX"] == 2].copy()
print("Female sample size::", df_female.shape)
```

## Basic cleaning

```python
# Convert invalid codes
df_clean = df_female.copy()
invalid_codes = {
    # targets
    "MENTHLTH":[77, 99],
    "GENHLTH":[7, 9],

    # demographic
    "_RACE1": [9],
    "EDUCA": [9],
    "INCOME3": [77, 99],
    "MARITAL": [9],
    "EMPLOY1": [9],
    "CHILDREN": [99],

    # behavioral
    "EXERANY2": [7, 9],
    "SLEPTIM1": [77, 99],
    "SMOKE100": [7, 9],
    "ALCDAY4": [777, 999],
    "_BMI5": [9999],

    # chronic
    "DIABETE4": [7, 9],
    "CVDCRHD4": [7, 9],
    "CVDSTRK3": [7, 9],
    "ASTHMA3": [7, 9],
    "CHCCOPD3": [7, 9],
    "HAVARTH4": [7, 9],
    "CHCKDNY2": [7, 9],
    "ADDEPEV3": [7, 9],

    # healthcare access
    "PRIMINSR": [77, 99],
    "PERSDOC3": [7, 9],
    "MEDCOST1": [7, 9],
    "CHECKUP1": [7, 9],

}

# ACE
for col in needed["ACE"]:
  invalid_codes[col] = [7, 9]

# SDHE
for col in needed["SDHE"]:
  invalid_codes[col] = [7, 9]

# conversion
for col, codes in invalid_codes.items():
  if col in df_clean.columns:
    df_clean[col] = df_clean[col].replace(codes, np.nan)

print("Invalid code conversion completed.")
print("df_clean shape:", df_clean.shape)
```

## Deal with special valid code

```python
# MENTHLTH: 88 - indicates that there were 0 days in the past 30 days with poor mental health.
df_clean["MENTHLTH"] = df_clean["MENTHLTH"].replace(88, 0)

# CHILDREN: 88 - have no kids
df_clean["CHILDREN"] = df_clean["CHILDREN"].replace(88, 0)

# _BMI5 = BMI * 100
# convert to real bmi value
df_clean["BMI"] = df_clean["_BMI5"] / 100

print("Special value conversion completed.")
display(df_clean[["MENTHLTH", "CHILDREN", "_BMI5", "BMI"]].head())
```

## Convert ALCDAY

```python
def convert_alcday4(x):
  if pd.isna(x):
    return np.nan

  # 888 = No alcohol consumption in the past 30 days
  if x == 888:
    return 0

  # 101-107 = how many days have drink per week
  # eg. 101 = drink 1 day in the past 30 days
  if 101 <= x <= 107:
    return (x-100) * 30 / 7

  # 201-230
  # eg. 205 = drink 5 days in the past 30 days
  if 201 <= x <= 230:
    return x - 200

  return np.nan

df_clean["alcohol_days_30"] = df_clean["ALCDAY4"].apply(convert_alcday4)

print("ALCDAY4 converted to alcohol_days_30.")
print(df_clean["alcohol_days_30"].describe())
```

# build target feature

```python
# For RQ1 & RQ2
# target_fmd = 1 - frequent mental distress, MENTHLTH >= 14
df_clean["target_fmd"] = np.where(
    df_clean["MENTHLTH"] >= 14,
    1,
    np.where(df_clean["MENTHLTH"].notna(), 0, np.nan)
)

print("target_fmd distribution:")
print(df_clean["target_fmd"].value_counts(dropna=False))
print(df_clean["target_fmd"].value_counts(normalize=True, dropna=False))
```

```python
# For RQ3
# target_resilience = 1 - general health condiition is good (GENHLTH in 1, 2, 3)
# target_resilience = 0 - general health condition is poor (GENHLTH in 4, 5)
df_clean["target_resilience"] = np.where(
    df_clean["GENHLTH"].isin([1, 2, 3]),
    1,
    np.where(df_clean["GENHLTH"].isin([4, 5]), 0, np.nan)
)

print("\ntarget_resilience distribution:")
print(df_clean["target_resilience"].value_counts(dropna=False))
print(df_clean["target_resilience"].value_counts(normalize=True, dropna=False))
```

## Build chronic count

```python
# 1 = Yes, 2 = No
chronic_yesno = [
    "CVDCRHD4", "CVDSTRK3", "ASTHMA3",
    "CHCCOPD3", "HAVARTH4", "CHCKDNY2", "ADDEPEV3"
]

for col in chronic_yesno:
  if col in df_clean.columns:
    df_clean[col + "_bin"] = df_clean[col].map({
        1: 1,
        2: 0
    })

# For DIABETE4 1 = yes, 0 = no
# 1 = Yes
# 2 = Yes, but only during pregnancy
# 3 = No
# 4 = No, pre-diabetes / borderline diabetes
if "DIABETE4" in df_clean.columns:
  df_clean["DIABETE4_bin"] = df_clean["DIABETE4"].map({
      1: 1,
      2: 0,
      3: 0,
      4: 0
  })

chronic_bin_cols = [
    col + "_bin" for col in needed["chronic"]
    if col + "_bin" in df_clean.columns
]

df_clean["chronic_count"] = df_clean[chronic_bin_cols].sum(axis=1, min_count=1)
print("chronic_count created.")
print("Used columns:", chronic_bin_cols)
print(df_clean["chronic_count"].describe())
print(df_clean["chronic_count"].value_counts(dropna=False).sort_index())
```

The variable `chronic_count` was created by converting eight chronic condition variables into binary indicators and summing them for each respondent.

## Create ACE_score

```python
# create binary ACE value & ACE score
# 1 = yes, 2 = no
ace_yesno = [
    "ACEDEPRS", "ACEDRINK", "ACEDRUGS", "ACEPRISN"
]

for col in ace_yesno:
  if col in df_clean.columns:
    df_clean[col + "_bin"] = df_clean[col].map({
        1: 1,
        2: 0
    })

# ACEDIVRC
# 1 = yes, 2 = no, 8 = Parents not married
# consider 8 as no parental divorce/separation
if "ACEDIVRC" in df_clean.columns:
  df_clean["ACEDIVRC_bin"] = df_clean["ACEDIVRC"].map({
      1: 1,
      2: 0,
      8: 0
  })

# ACE variables of frequency
# 1 = never, 2 = once, 3 = more than once
ace_freq = [
    "ACEPUNCH", "ACEHURT1", "ACESWEAR", "ACETOUCH", "ACETTHEM", "ACEHVSEX"
]
for col in ace_freq:
  if col in df_clean.columns:
    df_clean[col + "_bin"] = df_clean[col].map({
        # 1 = yes, 0 = no
        1: 0,
        2: 1,
        3: 1
    })

# positive / care variables
# 1 = never
# 2 = a little of the time, 3 = some of the time, 4 = most of the time
# 5 = all of the times
# Less care is considered adversity
for col in ["ACEADSAF", "ACEADNED"]:
  if col in df_clean.columns:
    df_clean[col + "_bin"] = df_clean[col].map({
        1: 1,
        2: 1,
        3: 1,
        4: 0,
        5: 0
    })

# total num of ACE binary cols
ace_bin_cols = [
    col + "_bin" for col in needed["ACE"]
    if col + "_bin" in df_clean.columns
]
# Record how many ACE items each sample answered
df_clean["ACE_answered_count"] = df_clean[ace_bin_cols].notna().sum(axis=1)

# if all ace items are missing, set ACE_score as NaN
df_clean["ACE_score"] = df_clean[ace_bin_cols].sum(axis=1, min_count=1)

# high ace score: ACE_score >= 4
df_clean["ACE_high_4"] = np.where(
    df_clean["ACE_score"] >= 4,
    1,
    np.where(df_clean["ACE_score"].notna(), 0, np.nan)
)

print("ACE_score created.")
print("Used ACE binary columns:", ace_bin_cols)
print("\nACE_answered_count distribution:")
print(df_clean["ACE_answered_count"].value_counts(dropna=False).sort_index())

print("ACE_score summary:")
print("\nACE_score distribution:")
print(df_clean["ACE_score"].value_counts(dropna=False).sort_index())

print("\nACE_high_4 distribution:")
print(df_clean["ACE_high_4"].value_counts(dropna=False))
print(df_clean["ACE_high_4"].value_counts(normalize=True, dropna=False))
```

ACE_score was constructed by recoding 13 ACE items into binary adversity indicators and summing them for each respondent. Respondents with no valid ACE item responses were assigned missing ACE_score rather than 0, because absence of module data does not imply absence of childhood adversity. Among female respondents, 25,528 had a valid ACE_score, and 6,004 were classified as high ACE using the threshold ACE_score >= 4.

Q: Why 4?

A: https://www.cdc.gov/violenceprevention/aces/about.html
Following the original CDC-Kaiser ACE Study literature, high childhood adversity was defined as ACE_score >= 4. The CDC describes the ACE score as “the total sum of the different categories of ACEs reported by participants” and reports ACE score prevalence using a “4 or more” category. Felitti et al. also specifically analysed “persons who had experienced four or more categories of childhood exposure”, showing that this group had substantially increased adult health risks compared with those with no ACE exposure.

## Create SDHE burden

```python
# create SDHE burden: indicates the quantity of current social needs/living difficulties

# LSATISFY: life satisfaction
# 1 = very satisfied, 2 = satisfied
# 3 = dissatisfied, 4 = very dissatisfied
# 3 & 4 are considered burden
df_clean["LSATISFY_burden"] = df_clean["LSATISFY"].map({
    1: 0,
    2: 0,
    3: 1,
    4: 1
})

# EMTSUPRT: Emotional support
# 1 = Always, 2 = Usually
# 3 = Sometimes, 4 = Rarely, 5 = Never
# 3, 4, 5 considered as low support (burden)
df_clean["EMTSUPRT_burden"] = df_clean["EMTSUPRT"].map({
    1: 0,
    2: 0,
    3: 1,
    4: 1,
    5: 1
})

# SDHISOLT: Socail isolation
# 1 = Always, 2 = Usually, 3 = Sometimes
# 4 = rarely, 5 = never
# 1, 2, 3 considered isolation (burden)
df_clean["SDHISOLT_burden"] = df_clean["SDHISOLT"].map({
    1: 1,
    2: 1,
    3: 1,
    4: 0,
    5: 0
})

# yes / no type SDHE variables
# 1 = yes, 2 = no
# indicates the respondent has experienced this social difficulty
# 1 - burden

yes_no_sdhe = [
    "SDHEMPLY",   # emplyment difficulty
    "FOODSTMP",   # used as material hardship proxy
    "SDHBILLS",   # difficulty paying bills
    "SDHUTILS",   # utility shutoff threat
    "SDHTRNSP"    # transpotation barrier
]

for col in yes_no_sdhe:
  if col in df_clean.columns:
    df_clean[col + "_burden"] = df_clean[col].map({
        1: 1,
        2: 0
    })

# SDHFOOD1: food insecurity
# 1 = Always, 2 = Usually, 3 = Sometims
# 4 = rarely, 5 = never
# 1, 2, 3 considered as burden
df_clean["SDHFOOD1_burden"] = df_clean["SDHFOOD1"].map({
    1: 1,
    2: 1,
    3: 1,
    4: 0,
    5: 0
})

# SDHSTRE1: Stress
# 1 = Always, 2 = Usually, 3 = Sometims
# 4 = rarely, 5 = never
# 1, 2, 3 considered as burden
df_clean["SDHSTRE1_burden"] = df_clean["SDHSTRE1"].map({
    1: 1,
    2: 1,
    3: 1,
    4: 0,
    5: 0
})

sdhe_burden_cols = [
    "LSATISFY_burden", "EMTSUPRT_burden", "SDHISOLT_burden",
    "SDHEMPLY_burden", "FOODSTMP_burden", "SDHBILLS_burden",
    "SDHUTILS_burden", "SDHTRNSP_burden", "SDHFOOD1_burden",
    "SDHSTRE1_burden"
]
sdhe_burden_cols = [
    col for col in sdhe_burden_cols
    if col in df_clean.columns
]
print("Number of SDHE burden columns:", len(sdhe_burden_cols))
print("Used SDHE burden columns:", sdhe_burden_cols)
```

```python
# Record how many SDHE items each sample answered
df_clean["SDHE_answered_count"] = df_clean[sdhe_burden_cols].notna().sum(axis=1)

# If all SDHE items are missing, SDHE_burden remains NaN
df_clean["SDHE_burden"] = df_clean[sdhe_burden_cols].sum(axis=1, min_count=1)
print("SDHE_burden created.")

print("\nSDHE_answered_count distribution:")
print(df_clean["SDHE_answered_count"].value_counts(dropna=False).sort_index())

print("\nSDHE_burden summary:")
print(df_clean["SDHE_burden"].describe())

print("\nSDHE_burden distribution:")
print(df_clean["SDHE_burden"].value_counts(dropna=False).sort_index())
```

SDHE_burden was constructed by converting 10 SDHE items into binary burden indicators and summing them for each respondent. Respondents with no valid SDHE item responses were assigned missing SDHE_burden rather than 0, because absence of module data does not imply absence of social needs. Among female respondents, 135,322 had a valid SDHE_burden score. The score ranged from 0 to 10, with a mean of approximately 1.50.

## Create ACE x SDHE

```python
# create ACE x SDHE
df_clean["ACE_x_SDHE"] = df_clean["ACE_score"] * df_clean["SDHE_burden"]
print("ACE_x_SDHE created.")
print(df_clean["ACE_x_SDHE"].describe())
print(df_clean["ACE_x_SDHE"].value_counts(dropna=False).sort_index())
```

The interaction term `ACE_x_SDHE` was created by multiplying `ACE_score` and `SDHE_burden`. It is only available for respondents with both valid ACE and SDHE module data. In the female sample, 10,434 respondents had a valid interaction value, reflecting the overlap between the two optional BRFSS modules. The interaction term ranges from 0 to 120 in the cleaned data.

Q: Why ACE x SDHE?

A:

1. Interaction analysis is a recognised modelling approach for examining whether the joint effect of two exposures differs from what would be expected based on their individual effects.(VanderWeele & Knol, interaction tutorial)
2. Oregon State University's open epidemiology textbook explains:
Effect modification, also known as interaction, indicates that the relationship between exposure and outcome differs at different levels of a third variable. (Foundations of Epidemiology, effect modification)
3. The CDC explains that ACEs affect lifelong health and opportunities, noting their association with health outcomes such as chronic disease, behavioral health, and depression in adulthood. The CDC also mentions that differences in ACE risk are related to social and economic circumstances.
4. The product term ACE_score × SDHE_burden was included as a baseline interaction feature. This does not assume that the true relationship is strictly multiplicative; rather, it follows standard modelling practice for representing potential interaction or effect modification between two continuous predictors. More complex nonlinear relationships may be captured by tree-based or neural models, and the usefulness of the interaction term is evaluated empirically through model comparison.



# RQ1 dataset

```python
# RQ1 uses traditional baseline features only
# It should not be filtered by ACE or SDHE availability

rq1_features = (
    needed["demographic"] + ["EXERANY2", "SLEPTIM1", "SMOKE100", "alcohol_days_30", "BMI"]
    + needed["chronic"] + ["chronic_count"] + needed["healthcare_access"]
)

rq1_features = list(dict.fromkeys([col for col in rq1_features if col in df_clean.columns]))
df_rq1 = df_clean[
    df_clean["target_fmd"].notna()
][rq1_features + ["target_fmd"]].copy()

print("RQ1 dataset created.")
print("RQ1 dataset shape:", df_rq1.shape)

print("\nRQ1 target distribution:")
print(df_rq1["target_fmd"].value_counts(dropna=False))
print(df_rq1["target_fmd"].value_counts(normalize=True, dropna=False))
```

Balanced accuracy

# RQ2 Dataset

```python
# RQ2 uses baseline + ACE + SDHE features
# requires valid target_fmd, ACE_score, SDHE_burden

rq2_features = (
    rq1_features + needed["ACE"] + needed["SDHE"]
    + ["ACE_score", "SDHE_burden", "ACE_x_SDHE"]
)

rq2_features = list(dict.fromkeys([col for col in rq2_features if col in df_clean.columns]))
df_rq2 = df_clean[
    df_clean["target_fmd"].notna()
    & df_clean["ACE_score"].notna()
    & df_clean["SDHE_burden"].notna()
][rq2_features + ["target_fmd"]].copy()

print("RQ2 dataset created.")
print("RQ2 dataset shape:", df_rq2.shape)

print("\nRQ2 target distribution:")
print(df_rq2["target_fmd"].value_counts(dropna=False))
print(df_rq2["target_fmd"].value_counts(normalize=True, dropna=False))
```

# RQ3 Dataset

```python
# RQ3 focuses on high-ACE (>=4) women and predicts resilience
# require valid SDHE_burden

rq3_features = (
    needed["SDHE"]
    + ["SDHE_burden"]
    + needed["healthcare_access"]
    + ["EXERANY2", "SLEPTIM1", "SMOKE100", "alcohol_days_30", "BMI"]
    + needed['demographic'] + ["chronic_count"]
)

rq3_features = list(dict.fromkeys([col for col in rq3_features if col in df_clean.columns]))
df_rq3 = df_clean[
    (df_clean["ACE_score"] >= 4) & (df_clean["target_resilience"].notna())
][rq3_features + ["target_resilience", "ACE_score"]].copy()

print("RQ3 dataset created.")
print("RQ3 dataset shape:", df_rq3.shape)
print("\nRQ3 target distribution:")
print(df_rq3["target_resilience"].value_counts(dropna=False))
print(df_rq3["target_resilience"].value_counts(normalize=True, dropna=False))
```
